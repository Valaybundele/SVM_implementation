```
from cvxopt import matrix as cvxopt_matrix
from cvxopt import solvers as cvxopt_solvers
y_train = y_train.reshape(-1,1)*1. 
```

```
y_test = y_test.rename_axis('ID').values*1.

print(y_test)
```

```
y_test[0]
```

```
y_train
```

```
C = 10
```

```
X_dash=y_train*x_train*1.
```

```
import numpy as np
H = np.dot((X_dash),(np.transpose(X_dash)))*1.
```

```
H
```

```
P = cvxopt_matrix(H)
q = cvxopt_matrix(-np.ones((460, 1)))
G = cvxopt_matrix(np.vstack((np.eye(460)*-1,np.eye(460))))
h = cvxopt_matrix(np.hstack((np.zeros(460), np.ones(460) * C)))
A = cvxopt_matrix(y_train.reshape(1, -1))
b = cvxopt_matrix(np.zeros(1))
```

```
cvxopt_solvers.options['show_progress'] = False
cvxopt_solvers.options['abstol'] = 1e-10
cvxopt_solvers.options['reltol'] = 1e-10
cvxopt_solvers.options['feastol'] = 1e-10
```

```
sol = cvxopt_solvers.qp(P, q, G, h, A, b)
alphas = np.array(sol['x'])
```

```
w = ((y_train * alphas).T @ x_train).reshape(-1,1)
S = (alphas > 1e-4).flatten()
b = y_train[S] - np.dot(x_train[S], w)

#Display results
print('Alphas = ',alphas[alphas > 1e-4])
print('w = ', w.flatten())
print('b = ', b[0])
```
